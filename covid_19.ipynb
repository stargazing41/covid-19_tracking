```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nytimes_dat = 'https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-states.csv'
dat = pd.read_csv(nytimes_dat, error_bad_lines=False)
```

```python
dat.shape
```

```python
dat
```

```python
states=[]
for d in range (dat.shape[0]):
    if dat.iloc[d,1] not in states:
        states.append(dat.iloc[d,1])
```

```python
for s in range (len(states)):
    dat_state = []
    for i in range(dat.shape[0]):
        if dat.iloc[i,1] == states[s]:
            dat_state.append(dat.iloc[i, 3])
    plt.plot(dat_state)
    plt.ylabel('Test Positive Cases')
    plt.title(states[s])
    plt.show()
```
